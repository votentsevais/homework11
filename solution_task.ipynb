# Задание №1
Определите какому фильму было выставлено больше всего оценок 5.0.

```python
movies= pd.read_csv('movies.csv')
ratings = pd.read_csv('ratings.csv')
```

```python
## Отсортировываем фильмы только с рейтингом 5.0. 
ratings=ratings[ratings.rating == 5.0]
##Затем считаем количество повторений фиильмов с этой оценкой. 
best_movie = ratings['movieId'].value_counts().index.tolist()[0]
```

```python
##Таким образом поучаем самый популярный
movies[movies.movieId == best_movie]['title']
```

# Задание №2
По данным файла power.csv посчитайте суммарное потребление стран Прибалтики (Латвия, Литва и Эстония) категорий 4, 12 и 21 за период с 2005 по 2010 года. Не учитывайте в расчетах отрицательные значения quantity.

```python
power= pd.read_csv('power.csv')
power = power[((power.country == 'Latvia') | (power.country == 'Lithuania') | (power.country == 'Estonia'))
      & (2005 <= power.year) & (power.year<=2010)
     & ((power.category == 4.0) | (power.category == 12.0) | (power.category == 21.0))
   & (power.quantity >= 0)].sort_values('quantity', ascending=True)
```

```python
power['quantity'].sum()
```
